# Q-Q Plot

```python
#Q-Q plot
import numpy as np 
import pylab 
import scipy.stats as stats

# N(0,1)
std_normal = np.random.normal(loc = 0, scale = 1, size=100)
print(std_normal)
# 0 to 100th percentiles of std-normal
for i in range(0,101):
    print(i, np.percentile(std_normal,i))


```

```python
# generate 100 sanples from N(20,5)
def plot_qq(size=10):
    
    plt.figure(figsize=(12,6))
    ax = plt.subplot(1,2,1)
    measurements = np.random.normal(loc = 0, scale = 1, size=size) 
    count, bins, ignored = plt.hist(measurements, 10, density=True)
    pylab.tick_params(axis='both',color='c',labelcolor='c',labelsize='large')
    plt.title("Histogram of measurements with 10 bins", color='w', size='x-large')
#     plt.show()
#     ax.set_title('Histogram')
#     ax.set_xlabel('Value bin')
#     ax.set_ylabel('Frequency')

    # plt.xlim(-1,1)
    plt.subplot(1,2,2)
    stats.probplot(measurements, dist="norm", plot=pylab)
    pylab.tick_params(axis='both',color='c',labelcolor='c',labelsize='large')
    pylab.xlabel(xlabel="Theoritical Quantities", color="w")
    pylab.ylabel(ylabel="Ordered Values", color="w")
    plt.show()
```

```python
# generate 100 sanples from N(20,5)
def plot_qq(size=10):
    measurements = np.random.normal(loc = 0, scale = 1, size=size) 
    #try size=1000
    print("min {0}, max {1}".format(min(measurements),max(measurements)))


    # plt.xlim(-1,1)
    count, bins, ignored = plt.hist(measurements, 10, density=True)
    pylab.tick_params(axis='both',color='c',labelcolor='c',labelsize='large')
    plt.title("Histogram of measurements with 10 bins", color='w', size='x-large')
    plt.show()

    # plt.xlim(-1,1)
    stats.probplot(measurements, dist="norm", plot=pylab)
    pylab.tick_params(axis='both',color='c',labelcolor='c',labelsize='large')
    pylab.xlabel(xlabel="Theoritical Quantities", color="w")
    pylab.ylabel(ylabel="Ordered Values", color="w")
    pylab.show()
```



###### with np.random.normal (loc = 0, scale = 1, size=5)

```python
plot_qq(500)
```

###### with np.random.normal (loc = 0, scale = 1, size=10000)

```python
plot_qq(10000)
```
