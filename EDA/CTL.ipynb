```python
#Q-Q plot
import numpy as np 
import pylab 
import scipy.stats as stats
from numpy.random import seed
from numpy.random import randint
from numpy import mean,std
from matplotlib import pyplot
# seed the random number generator
seed(1)
```

#### source - https://machinelearningmastery.com/a-gentle-introduction-to-the-central-limit-theorem-for-machine-learning/

```python
def plot_qq(sample,sample_count,sample_size):
    import matplotlib.pyplot as plt
    fig = plt.figure(figsize=(12,6))
    fig.suptitle('Sample Count {} with each sample of size {}'.format(sample_count,sample_size), 
             fontsize=20, color='y')
    
    ax = plt.subplot(1,2,1)
    count, bins, ignored = plt.hist(sample)
    pylab.tick_params(axis='both',color='c',labelcolor='c',labelsize='large')
#     plt.title("Histogram of measurements with 10 bins", color='w', size='x-large')
#     plt.show()
    ax.set_title('Histogram',color='w')
    ax.set_xlabel('Value bin',color='w')
    ax.set_ylabel('Frequency',color='w')

    # plt.xlim(-1,1)
    ax2 = plt.subplot(1,2,2)
    stats.probplot(sample, dist="norm", plot=pylab)
    pylab.tick_params(axis='both',color='c',labelcolor='c',labelsize='large')
    pylab.xlabel(xlabel="Theoritical Quantities", color="w")
    pylab.ylabel(ylabel="Ordered Values", color="w")
    ax2.set_title('Q Q',color='w')
    plt.show()
```

```python
# seed the random number generator
seed(1)
# calculate the mean of 60 dice rolls 1000 times
x_bar_60 = [mean(randint(1, 7, 50)) for _ in range(1000)]
print("The Mean of the Sample Mean --->{0}".format(mean(x_bar_60)))
print("Here the Mean of the Sample Mean is much closer to the Actual/Population Mean ; (1 + 2 + 3 + 4 + 5 + 6) / 6 = 21/6 = 3.5")
# plot a Q-Q graph with x_bar (sample mean with each sample size of 60)
plot_qq(x_bar_60,sample_count=1000,sample_size=50)
```

```python
# seed the random number generator
seed(1)
# calculate the mean of 10 dice rolls 1000 times
x_bar_10 = [mean(randint(1, 7, 10)) for _ in range(1000)]
# plot the distribution of sample means
print("The Mean of the Sample Mean --->{0}".format(mean(x_bar_10)))
print("Here the Mean of the Sample Mean differs a bit from  the Actual/Population Mean ; (1 + 2 + 3 + 4 + 5 + 6) / 6 = 21/6 = 3.5")
plot_qq(x_bar_10,sample_count=1000,sample_size=10)
```

```python
# seed the random number generator
seed(1)
# generate a sample of die rolls
xi_bar = randint(1, 7, 50)
print("The Mean of the Sample Mean --->{0} Std-dev -->{1}".format(mean(xi_bar),std(xi_bar)))
print("Here the Mean of the Sample Mean differs a lot from the Actual/Population Mean ; (1 + 2 + 3 + 4 + 5 + 6) / 6 = 21/6 = 3.5")
plot_qq(xi_bar,sample_count=1,sample_size=50)
```
